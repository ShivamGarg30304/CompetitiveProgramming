## FacetGrid

```python
# prompt: Using dataframe tips: Create 5 good facetgrids on tips dataset

import matplotlib.pyplot as plt
# FacetGrid based on sex and time
g = sns.FacetGrid(tips, col="sex", row="time")
g.map(plt.hist, "total_bill")

# FacetGrid based on smoker and day
g = sns.FacetGrid(tips, col="smoker", row="day")
g.map(plt.hist, "total_bill")

# FacetGrid based on sex and smoker
g = sns.FacetGrid(tips, col="sex", row="smoker")
g.map(plt.hist, "total_bill")

```

```python
tips = sns.load_dataset('tips')
g = sns.FacetGrid(tips, col='time')
```

```python
g = sns.FacetGrid(tips, col='time')
g.map(sns.histplot, 'tip', bins=range(0, 10, 1))
```

```python
g = sns.FacetGrid(tips, col='sex', hue='smoker')
g.map(sns.scatterplot, "total_bill", 'tip', alpha=0.7)
g.add_legend()
```
